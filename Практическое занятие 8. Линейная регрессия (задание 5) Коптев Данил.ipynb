```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import seaborn as sns
import math

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
```

# –ó–∞–¥–∞–Ω–∏–µ 5 

1) –°–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å–Ω—ã–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –∏–∑ 200 –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è –∑–∞–¥–∞—á–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –æ–¥–Ω–æ–π —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –∏ —á–µ—Ç—ã—Ä—å–º—è –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã–º–∏ —è–≤–ª—è—é—Ç—Å—è  –¥–≤–∞.  –ü–∞—Ä–∞–º–µ—Ç—Ä,  –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∏–π  —Å—Ç–µ–ø–µ–Ω—å  —Ä–∞—Å—Å–µ—è–Ω–Ω–æ—Å—Ç–∏  –¥–∞–Ω–Ω—ã—Ö,  —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ä–∞–≤–Ω—ã–º  5.  –û–±–µ—Å–ø–µ—á–∏–º  –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç—å  —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤,  –∑–∞–¥–∞–≤  –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–º—É –ø–∞—Ä–∞–º–µ—Ç—Ä—É.

```python
data, target, coef = make_regression(n_samples=200, n_features=4, n_informative=2, noise=5, coef=True, random_state=25)
```

2) –í—ã–ø–æ–ª–Ω–∏–º —Ä–∞–∑–±–∏–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 70/30

3) –ü–æ—Å—Ç—Ä–æ–∏–º  –º–æ–¥–µ–ª–∏  –ª–∏–Ω–µ–π–Ω–æ–π  —Ä–µ–≥—Ä–µ—Å—Å–∏–∏,  –∏—Å–ø–æ–ª—å–∑—É—é—â–∏–µ ùêø2-–∏ ùêø1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ç–æ—Ä—ã. –û–±—É—á–∏—Ç—å  —ç—Ç–∏  –º–æ–¥–µ–ª–∏  –Ω–∞  –æ–±—É—á–∞—é—â–µ–π  –≤—ã–±–æ—Ä–∫–µ,  –∏—Å–ø–æ–ª—å–∑—É—è  —Å–ª–µ–¥—É—é—â–∏–µ  –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏: 0,001;0,01;0.2;1;2;10.

4) –î–ª—è –∫–∞–∂–¥–æ–π –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –æ–±—ä–µ–∫—Ç–∞—Ö —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏. 

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫—É—é –æ—à–∏–±–∫—É –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

5) –í—ã–≤–µ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –ø–æ—Å–ª–µ –æ–±—É—á–µ–Ω–∏—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏. 

–ó–∞–ø–∏—à–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.

6) –í—ã–ø–æ–ª–Ω–∏–º  –∞–Ω–∞–ª–∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö  —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, random_state=25 ,test_size=0.3)

regul_coefs= [0.001,0.01,0.2,1,2,10]

for regul_coef in regul_coefs:

  print("\n–î–ª—è –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–∞ —Ä–µ–≥—É–ª—è—Ä–∞–∑–∞—Ü–∏–∏", regul_coef)

  L1 = Lasso(regul_coef)
  L2 = Ridge(regul_coef)
  
  L1.fit(x_train, y_train)
  L2.fit(x_train, y_train)

  L1_predict = L1.predict(x_test)
  L2_predict = L2.predict(x_test)

  print("L1")
  print("MSE", np.round(mean_squared_error(y_test, L1_predict),3))
  print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏', np.round(L1.coef_, 3))
  print(f"–£—Ä–æ–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:  y = x1*{ np.round(L1.coef_[0], 3)} + x2*{np.round(L1.coef_[1], 3)} " + 
        f" + x3*{np.round(L1.coef_[2],3)} + x4*{np.round(L1.coef_[3],3)} + {np.round(L1.intercept_,3)}")

  print("L2")
  print("MSE", mean_squared_error(y_test, L2_predict))
  print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏', L2.coef_)
  print(f"–£—Ä–æ–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:  y = x1*{ np.round(L2.coef_[0], 3)} + x2*{np.round(L2.coef_[1], 3)} " + 
        f" + x3*{np.round(L2.coef_[2],3)} + x4*{np.round(L2.coef_[3],3)} + {np.round(L2.intercept_,3)}")
```

2. –ü—Ä–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–º –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–µ —Ä–µ–≥—É–ª—è—Ä–∞–∑–∞—Ü–∏–∏ $MSE$ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –Ω–∞–∏–ª—É—á—à–∏–º, –Ω–æ –∏ –ø—Ä–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–µ —Ä–µ–≥—É–ª—è—Ä–∞–∑–∞—Ü–∏–∏ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –Ω–∞–∏–±–æ–ª–µ–µ –ø—Ä–µ–¥–ø–æ—á–∏—Ç–µ–ª—å–Ω—ã–º –∏–∑ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —á–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–∞ —Ä–µ–≥—É–ª—è—Ü–∏–∏, —Ç–µ–º –Ω–∏–∂–µ —Å–ª–æ–∂–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏, —á–µ–º –º–µ–Ω—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–∞ —Ä–µ–≥—É–ª—è—Ü–∏–∏, —Ç–µ–º –≤—ã—à–µ —Ä–∏—Å–∫ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è, –ø–æ—ç—Ç–æ–º—É –Ω—É–∂–µ–Ω –±–∞–ª–∞–Ω—Å. –ù–∞–ø—Ä–∏–º–µ—Ä, $L_2$ –ø—Ä–∏ –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç–µ —Ä–µ–≥—É–ª—è—Ü–∏–∏  $0.2$ –ø–æ–∫–∞–∑–∞–ª –Ω–∞–∏–ª—É—á—à–µ–µ $MSE$.

3. –¢–∞–∫ –∂–µ –º—ã –º–æ–∂–µ–º –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ –ø–æ—Å–ª–µ –µ–≥–æ –∫–æ—ç—Ñ–∏—Ü–µ–Ω—Ç—Ä–∞ —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ $0.2$ –Ω–µ–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã–µ –≤–µ—Å–∞ $L_1$ —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è —Ä–∞–≤–Ω—ã–º–∏ –Ω—É–ª—é, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ $L_1$-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ç–æ—Ä –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å –æ—Ç–±–æ—Ä –Ω–∞–∏–±–æ–ª–µ–µ –∑–Ω–∞—á–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
