Используя инструментарий модуля datasets, импортировать встроенный набор данных iris(часто используется авторами учебников и учебных курсов по машинному обучению в качестве учебного набора данных; содержит информацию о 150 объектах –цветах ириса, которые на основании значений четырех признаков необходимо отнести к одному из трех сортов).О наборе данных.
```python
data = iris.data
data
```

6. Выполнить  разбиение  набора  данных  (значения  признаков  и  метки  классов)  на обучающую и тестовую выборки в соотношении 70/30.

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=25)
```

7. Создать модель линейной классификации SGDClassifierи вывести список доступных параметров модели.

```python
SGDC = SGDClassifier()
SGDC.get_params()
```

8. Создать  сетку  параметров,  включающую  4  вида  функции  потерь,  два  типа регуляризаторов, 5 значений коэффициента регуляризации от 0.0001 до 0.001, и число итераций от 5 до 10 с шагом 1.

9. Создать объект GridSearchCV, передать ему созданный ранее классификатор и сетку параметров и обучить его. В качестве метрики использовать accuracy. Предусмотреть вывод  времени,  затраченного  на  перебор  по  сетке  (можно  использовать команду%%time).

```python
%%time

gridSearch = GridSearchCV(
    SGDC, 
    {
    'loss': ['log_loss', 'squared_hinge', 'hinge', 'modified_huber'],
    'penalty': ['l1', 'l2'],
    'alpha': np.linspace(0.0001, 0.001, 5),
    'n_iter_no_change': np.arange(5,11)
    }, 
    cv=5, 
    scoring='accuracy')
gridSearch.fit(x_train, y_train)
```

10. Вывести  лучший  классификатор,  лучший  набор  параметров  и  оценку  лучшего классификатора в соответствии с заданной метрикой. Получить оценки алгоритма по последним 10 наборам параметров.

```python
print("Лучший классификатор: ", gridSearch.best_estimator_, "\n")
print("Лучший набор параметров: ", gridSearch.best_params_, "\n")
print("Оценка лучшего классификатора: ", gridSearch.best_score_, "\n")
print("Оценки алгоритма по последним 10 наборам параметров",  gridSearch.cv_results_['mean_test_score'][:10])
```

11. Проанализировать полученные результаты, сделать выводы.

Для его достижения был использован регуляризаторо $L_1$ коэфицентом регуляризации 0.000775, с показатель остановки  9, а функция потерь была использована hinge. Общее время реализации алгоритма 4.46 s.

13. Вывести те же показатели, что и в п. 10. Сопоставить полученные результаты и время на перебор с результатами, полученными при выполнении п. 10.

```python
%%time

randomizedSearch = RandomizedSearchCV(
    SGDC, 
    {
    'loss': ['hinge', 'log_loss', 'modified_huber', 'squared_hinge'],
    'penalty': ['l1', 'l2'],
    'alpha': np.linspace(0.0001, 0.001, 5),
    'n_iter_no_change': np.arange(5,11)
    }, 
    n_iter=50, 
    cv=5, 
    scoring='accuracy', 
    random_state=25)
randomizedSearch.fit(x_train, y_train)
```

```python
print("Лучший классификатор: ", randomizedSearch.best_estimator_, "\n")
print("Лучший набор параметров: ", randomizedSearch.best_params_, "\n")
print("Оценка лучшего классификатора: ", randomizedSearch.best_score_, "\n")
print("Оценки алгоритма по последним 10 наборам параметров", randomizedSearch.cv_results_['mean_test_score'][:10])
```
