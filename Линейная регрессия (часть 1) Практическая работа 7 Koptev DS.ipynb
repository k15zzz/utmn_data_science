**Задание 2**

1. Сгенерируем  модельный  набор  данных  для  задачи  линейной  регрессии  с  одной целевойпеременной и двумя признаками, из которых информативным является только один. 

```python
data, target, coef = make_regression(n_samples=200, n_features=2, n_informative=1, noise=5, coef=True, random_state=25)
print(data, target, coef)
```

2. Выведим на одном графике сгенерированный набор данных (облако точек в координатах информативный  признак –целевая  переменная)  и  прямую  регрессии  со сгенерированными коэффициентами.

```python
x = data[:,0]
y = target

z = np.polyfit(x, y, 1 )
p = np.poly1d(z)
plt.scatter(x = x, y = y, label="Облако точек")
plt.xlabel("Информативный признак")
plt.ylabel("Целевой признак")
plt.plot (x, x*coef[0], color="red", label="Прямая регрессия") 
plt.legend()
plt.title("Графике сгенерированных наборов данных и прямой  регрессии с шумом 5")
plt.show()
```

3. Поэкспериментируем  с  величиной  шума:  задать  значения  соответствующего параметра равными 7, 10 и 15, вывести графики (вместе с прямой регрессии) в одном ряду,  с  заголовками,  сообщающими  об  используемом  значении  параметра  шума. Прокомментировать полученные результаты.

4. Выполним  разовое  разбиение  всех  полученных  наборов  данных  (с  различными значениями шума) на обучающую и тестовую выборки в соотношении 70/30.

5. Для каждого набора создаем модель линейной регрессии LinearRegressionи обучить ее на обучающей выборке.

6. Получим предсказания обученных моделей для объектов тестовых выборок. Вывести массивы ответов на тестовых выборках и массивы предсказанных моделью значений. Представим  результаты  в  графической  форме  и  проанализировать  их. Прокомментировать полученные результаты.

7. Выведим коэффициенты уравнений, сформированных при генерации наборов данных, и коэффициенты регрессии, полученные после обучения моделей. 

8. Запишем  уравнения  исходных  зависимостей  (полученные  при  генерации  наборов данных)  и  полученные  уравнения  регрессии.  Внимание:  не  следует  забывать  про свободный член в уравнении регрессии!

```python
noizes = [7, 10, 15]

for noise in noizes :
  # unit 3
  data, target, coef = make_regression(n_samples=200, n_features=2, n_informative=1, noise=noise, coef=True, random_state=25)

  x = data[:,0]
  y = target

  z = np.polyfit(x, y, 1 )
  p = np.poly1d(z)
  plt.scatter(x = x, y = y, label="Облако точек")
  plt.xlabel("Информативный признак")
  plt.ylabel("Целевой признак")
  plt.plot (x, x*coef[0], color="red", label="Прямая регрессия") 
  plt.legend()
  plt.title("Графике сгенерированных наборов данных и прямой  регрессии с шумом " + str(noise))
  plt.show()

  # unit 4
  x_train, x_test, y_train, y_test = train_test_split(data, target, random_state=25 ,test_size=0.3)

  # unit 5
  linearRegression = LinearRegression()
  linearRegression.fit(x_train, y_train)

  # unit 6
  predict = linearRegression.predict(x_test)
  print(predict)

  z = np.polyfit(x, y, 1 )
  p = np.poly1d(z)
  plt.xlabel("Информативный признак")
  plt.ylabel("Целевой признак")
  plt.scatter(x = x_test[:,0], y = y_test, label="Значения из тестовой выборки")
  plt.scatter(x = x_test[:,0], y = predict, label="Предсказанные значения") 
  plt.legend()
  plt.title("Графике массива ответа на тестовых выборках и массива предсказанных моделью значений с шумом " + str(noise))
  plt.show()

  # unit 7 
  print('Коэффициенты построенного уравнения регрессии', linearRegression.coef_)
  print('Коэффициенты уравнения', coef)

  print('Свободный член: ', linearRegression.intercept_)

  # unit 8
  print(f"Уровнение регрессии:  y = x1 * {linearRegression.coef_[0]} + x2 * {linearRegression.coef_[1]} + {linearRegression.intercept_}")
  print(f"Уравнение исходной зависимости: y = {coef[0]} * x1 + {coef[1]} * x2")

  print('-----------------------------------------------------------------')
```
